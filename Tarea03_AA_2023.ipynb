### **Matriz de Correlación**

La correlación de Pearson es una medida estadística que expresa el nivel o grado en que dos variables estan linealmente relacionadas. El rango de valores que toma la correlación se encuentra entre -1 a 1. [Más detalle](https://github.com/VicoSan07/AprendizajeAutomatico/blob/main/Actividad1_AA_2023.ipynb)

Cuando contamos con un conjunto mayor o igual a 3 variables, es posible definir una **matriz de correlación**.

```python
# Matriz de correlación de vinos rojos
dfwiner.corr()
```

Notemos que la mayoría de variables mantienen correlaciones a lo sumo moderadas con otras variables. Aún así notemos que existen relaciones estrechas como lo son

1. Ácido fijo y ácido cítrico, una relación clara teniendo en cuenta que el ácido citrico es uno de los 3 compuestos del ácido fijo.
2. Ácido fijo y pH, siendo que por definición el pH es un indicador de acidez
3. Dióxido de azufre libre y dioxido de azufre total, siendo uno de otro, complemento.

Otras relaciones un poco curiosas pero interesantes son aquellas referentes al porcentaje de alcohol, dentro de sus correlaciones 'más altas' están las correspondientes a la densidad del vino (nivel de agua) y calidad del mismo.

### **Prueba de Hipótesis**

Mencionaba que es un poco curiosa la relación que existe entre el porcentaje de alcohol y densidad de un vino debido a que su correlación es muy cercana a 0.5, encontrandose cercana al umbral de ser considerada una relación moderada entre ambas variables. 

Así que para no obviar que sí existe relación entre ambas variables desarrollaremos una prueba de hipotesis para correlación: H0: p=0 vs HA:p<>0

```python
from scipy.stats import pearsonr

#Coeficiente de correlación y p-valor de t
pearsonr(dfwiner.alcohol, dfwiner.density)
```

Debido a que el p-valor es menor a alpha de 0.05, se rechaza la Hipotesis nula, de manera que el coeficiente de correlación es significativamente distinto de 0.

### **Representación gráfica**

De esta último conjunto de variables visualmente podemos observar como la relación entre el porcentaje de alcohol y nivel de densidad de los vinos parece no ser lineal. Así como nuestra afirmación de que las correlaciones entre variables son a lo sumo moderadas entre las mismas.

```python
# Scatterplot
import seaborn as sns
sns.scatterplot(x="density", y="alcohol", data=dfwiner)
```
