### **Carga de Datasets**

```python
#Pandas es una librer铆a especializada en el manejo y an谩lisis de datos
import pandas as pd

#Carga de dataset 'vino rojo' 
dfwiner = pd.read_csv(r'C:/Users/vicos/Documents/winequality-red.csv',delimiter=";",
                      #Tipo de datos en cada columna
                      dtype={
                          'fixed acidity': float,
                          'volatile acidity': float,
                          'citric acid': float,
                          'residual sugar': float,
                          'chlorides': float,
                          'free sulfur dioxide': float,
                          'total sulfur dioxide': float,
                          'density': float,
                          'pH': float,
                          'sulphates': float,
                          'alcohol': float,
                          'quality': int,
                      })

#Carga de dataset 'vino blanco' 
dfwinew = pd.read_csv(r'C:/Users/vicos/Documents/winequality-white.csv',delimiter=";",
                      #Tipo de datos en cada columna
                      dtype={
                          'fixed acidity': float,
                          'volatile acidity': float,
                          'citric acid': float,
                          'residual sugar': float,
                          'chlorides': float,
                          'free sulfur dioxide': float,
                          'total sulfur dioxide': float,
                          'density': float,
                          'pH': float,
                          'sulphates': float,
                          'alcohol': float,
                          'quality': int,
                      })

dfwinew
```

### **Manipulaci贸n de datos**

Nuestro principal objetivo ser谩 unificar ambos datasets en uno solo, para ello, crearemos una columna adicional en el que identifiquemos el tipo de vino con el que se trabaja: blanco o rojo 

```python
# Crear columnas con string constante en cada fila
dfwiner['type']='red'
dfwinew['type']='white'
```

```python
# Visualizaci贸n del resultado
dfwiner[["quality", "type"]].head()
```

```python
# Concatenar los dos datasets en uno solo
dfwineall = pd.concat([dfwiner,dfwinew],ignore_index=True)
dfwineall
```

```python
# Obtener el top 3 de vinos con m谩s porcentaje de alcohol

dfwineall.sort_values(['alcohol'], ascending=False).head(3)
```

```python
# Sobre esta misma variable, podemos crear grupos en base al nivel de alcohol en los vinos

# niveles a separar
galco = [7,9,12,15]

# clasificacion de los niveles
namegalco = ['ligero','medio','fuerte']

# nueva columna con la clasificacion de alcohol
dfwineall['clasif'] = pd.cut(dfwineall['alcohol'], galco, labels = namegalco)
dfwineall
```

```python
# Uso de conjuntos

# Creamos un conjunto respecto a la calidad de los vinos en funci贸n de su tipo: blanco o rojo
stwiner = set(dfwineall.loc[dfwineall['type']=='red','quality'])
stwinew = set(dfwineall.loc[dfwineall['type']=='white','quality'])

print('Relaciones entre vinos blancos y rojos')
print('Los vinos blancos y rojos comparten la(s) siguiente(s) puntuacion(es) entre s铆:', stwinew.intersection(stwiner))
print('El vino blanco no comparte la(s) siguiente(s) puntuacion(es) con el vino rojo:', stwinew.difference(stwiner))

# Contaremos cuantos vinos existen por puntuaci贸n o score
print('\nConteo de vinos por puntuaci贸n:')
for i in stwiner.union(stwinew):
    print('Existen ',len(dfwineall.loc[dfwineall['quality']==i,].index),' vinos con una punutaci贸n de', i)
```
